```python
import cv2
```

```python
import numpy as np
def print_img(img):
    plt.imshow(np.array(img))
    plt.show()
```

```python
def generate_certi(img):
    generated_image=img.copy()
    font=cv2.FONT_HERSHEY_SIMPLEX
    coordinates=(500,500)
    font_size=4
    font_color=(100,100,100)
    font_thickness=8
    name='Piyush Gupta'
    cv2.putText(generated_image, name, coordinates, font, font_size, font_color, font_thickness)
    return generated_image
```

```python
img=cv2.imread(r'C:\Users\PIYUSH\Desktop\python_20-04-21\Day08\CertificateTemplate.jpg')
plt.imshow(img)
plt.show()
```

```python
certi=generate_certi(img)
```

```python
print_img(certi)
```
