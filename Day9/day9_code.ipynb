```python
def get_movie_id(num=10):
    links_data = pd.read_csv(r"C:\Users\PIYUSH\Downloads\links.csv")
    movie_id=list(links_data['imdbId'])
    start_index=0
    end_index=start_index+num
    return movie_id[start_index:end_index]
```

```python
def scrape_movie_with_id(movie_id):
    url="https://www.imdb.com/title/tt{}/".format(str(movie_id).zfill(7))
    raw_data=requests.get(url)
    data_soup=bs4.BeautifulSoup(raw_data.text,'html')
    current_page=data_soup.find("script",attrs={"type": "application/ld+json"})
    print(current_page)
    
    
    
    
```

```python
def collect_movie_data(movie_id):
    scrape_movie_with_id(movie_id)
```

```python
ids=get_movie_id(20)
for x in ids:
    collect_movie_data(x)
    break
```
